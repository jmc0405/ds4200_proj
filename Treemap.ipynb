```python
import pandas as pd
import numpy as np
```

```python
music_data = pd.read_csv('cleaned_music_data.csv')
music_data
```

```python
import plotly.express as px

genres_data = music_data[['School', 'Top Genre (#1)', 'Top Genre (#2)', 'Top Genre (#3)']].copy()

genres_melted = genres_data.melt(id_vars=['School'], value_vars=['Top Genre (#1)', 'Top Genre (#2)', 'Top Genre (#3)'], var_name='Genre_Rank', value_name='Genre')

treemap_data = genres_melted.groupby(['School', 'Genre']).size().reset_index(name='Count')

fig = px.treemap(treemap_data, path=[px.Constant("all"), 'School', 'Genre'], values='Count',
                 title='Distribution of Top Genres by School',
                 color='Count', color_continuous_scale='RdBu')
fig.show()

snippet = fig.to_html(full_html=True, include_plotlyjs="cdn")
with open("treemap_chart.html", "w") as f:
    f.write(snippet)
```
